```python
# Preprocessing
# 1. resizing
# 2. flatten

import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize

target = []
images = []
flat_data = []

DATADIR = '/content/images'
CATEGORIES = ['animal dog','animal elephant','animal giraffe','animal horse', 'animal monkey','insects','reptile snake']

for category in CATEGORIES:
  class_num = CATEGORIES.index(category)#label encoding the values
  path = os.path.join(DATADIR,category) #create path to use all the images
  
  for img in os.listdir(path):
    img_array = imread(os.path.join(path,img))
    img_resized = resize(img_array,(100,100,3)) #normalizes the evalue to 0-1
    flat_data.append(img_resized.flatten())
    images.append(img_resized)
    target.append(class_num)

flat_data = np.array(flat_data)
target = np.array(target)
images = np.array(images)
```

```python
len(flat_data[0])
```

```python
unique,count = np.unique(target,return_counts= True )
plt.bar(CATEGORIES,count)
```

```python
#split data into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(flat_data,target,test_size=0.3,random_state=109)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid = [
              {'C':[1,10,100,1000],'kernel': ['linear']},
              {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']},
              ]

svc = svm.SVC(probability=True)
clf = GridSearchCV(svc,param_grid)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
#save the model using pickle library
import pickle
pickle.dump(clf,open('img_model.p','wb'))
```

```python
#deployment
%%writefile project-app.py
import streamlit as st
import numpy as np
from skimage.io import imread
from skimage.transform import resize
import pickle
from PIL import Image
st.title("Image Classifier using Machine learning")
st.subheader("upload the image")

model = pickle.load(open('img_model.p','rb'))

uploaded_file = st.file_uploader("choose an image..")
if uploaded_file is not None:
  img= Image.open(uploaded_file)
  st.image(img,caption="Image Uploaded")

  if st.button('PREDICT'):
    CATEGORIES = ['animal dog','animal elephant','animal giraffe','animal horse', 'animal monkey','insects','reptile snake']
    st.write('Result....')
    flat_data = []
    img = np.array(img)
    img_resized = resize(img,(100,100,3))
    flat_data.append(img_resized.flatten())
    flat_data = np.array(flat_data)
    y_out = model.predict(flat_data)
    y_out = CATEGORIES[y_out[0]]
    st.write(f'PREDICTED OUTPUT : {y_out}')
    q=model.predict_proba(flat_data)
    for index, item in enumerate(CATEGORIES):
      st.write(f'{item} : {q[0][index]*100}')
```
